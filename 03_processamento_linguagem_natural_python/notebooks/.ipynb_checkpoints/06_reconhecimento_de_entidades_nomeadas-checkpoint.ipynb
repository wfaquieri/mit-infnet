```python
import spacy

nlp = spacy.load("pt_core_news_lg")
```

```python
from tqdm.auto import tqdm 
tqdm.pandas()
```

```python
data['docs'] = data.body.progress_map(nlp)
```

```python

companies = []
for doc in data.docs:
    orgs = []
    for ent in doc.ents:
        
        if ent.label_ != "ORG":
            continue
        orgs.append(ent.text)
    orgs = list(set(orgs))
    if len(orgs) > 1:
        companies.append(orgs)
```

```python
edges = []
for nodes in companies:
    edges.extend([[x, y] for x, y in combinations(nodes, 2)])

edges = pd.DataFrame(edges, columns=['source', 'target'])
edges = edges.groupby(edges.columns.tolist(),as_index=False).size()
edges.columns = ['source', 'target', 'weight']
edges = edges.sort_values('weight', ascending=False)
edges
```

```python
import networkx as nx


graph = nx.from_pandas_edgelist(edges, 'source',  'target', ['weight'])
```

```python
nx.write_graphml(graph, "./companies.graphml")
```
