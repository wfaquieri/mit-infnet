<font color='gray'>
    
*Stemming* ou *lemmatization* são etapas de #textnormalization. Como podemos notar ao utilizar *stemming* reduzimos o token à uma forma raiz ou básica que, nesse caso, não precisa ser uma palavra existente no idioma ou presente em um dicionário. Uma vantagem de se utilizar a lematização em detrimento do *stemming*, é facilitar o entendimento do texto analisado, dos tópicos extraídos ou dos grupamentos, uma vez que o radical ou *lemma* extraído são conhecidos pelo idioma. Exemplo: conta, celular, controle, ficar.

## Criar modelos utilizando vetorização de textos baseado em Bag of Words<a name="part3"></a>

12. Quando adotamos uma estratégia frequentista para converter textos em vetores, podemos fazê-lo de diferentes maneiras. Mostramos em aula as codificações One-Hot, TF e TF-IDF. Explique a principal motivação em adotar TF-IDF frente as duas outras opções.

<font color='gray'>    
TF-IDF é apenas uma forma de medir a importância relativa dos tokens no texto, transformando um documento em uma lista de números. Isso é importante, pois muitas técnicas de NLP necessitam do texto "vetorizado". Diferentemente de outras estratégias para converter textos em vetores, o TF-IDF informa a **importância relativa** de um token em um documento. Se um token aparece muitas vezes em um documento e raras vezes em outro documento, então deve receber um peso maior, pois **caracteriza** esse documento. Por outras palavras, um peso TF-IDF alto para uma palavra em um documento pode implicar que a palavra é relativamente exclusiva desse documento específico. Essa espécie de particularidade da palavra não apenas em relação ao próprio documento, mas também a todos os outros documentos que compõe o corpus não é capturada pelas codificações One-Hot e TF.
</font>

## Criar modelos baseados em Word Embedding<a name="part4"></a>

18. Neste projeto, usamos TF-IDF para gerar os vetores que servem de entrada para o algoritmo de LDA. Quais seriam os passos para gerar vetores baseados na técnica de Doc2Vec?

<font color='gray'>
Como o próprio nome já sugere, o objetivo é gerar vetores não de palavras, mas de documentos. Se em word2vec uma alternativa seria utilizar Continuous Bag of Words (CBOW) ou o modelo skip-gram e definir o sliding window size e, com base nesse tamanho de janela, identificar a probabilidade condicional de observar a palavra de saída com base nas palavras ao redor (contexto). Em Doc2Vec utiliza-se variações desses modelos: PV-DBOW e PV-DM. Os pesquisadores Mikilov e Le (2014) estenderam a ideia de contexto aos parágrafos ou frases (documentos), adicionando um rótulo ou ID. Dada uma dimensionalidade, encontre representações vetoriais para palavras de modo que palavras com significados semelhantes tenham vetores semelhantes. A idéia básica para o treinamento é que palavras que ocorrem em contextos semelhantes têm significados semelhantes. Isso é chamado de hipótese distributiva. Mas, em vez de usar apenas palavras para prever a próxima palavra, eles adicionaram outro vetor de recursos, que é exclusivo do documento. Assim, ao treinar os vetores de palavras W, o vetor de documento D também é treinado e, ao final do treinamento, ele guarda uma representação numérica do documento.
</font>

19. Em uma versão alternativa desse projeto, optamos por utilizar o algoritmo de K-Médias para gerar os clusters (tópicos). Qual das abordagens (TF-IDF ou Doc2Vec) seria mais adequada como processo de vetorização? Justifique com comentários sobre dimensionalidade e relação semântica entre documentos.

<font color='gray'>
    
A abordagem utilizando Doc2Vec seria a mais adequada, pois TF-IDF, embora seja fácil de calcular, apresenta algumas desvantagens significativas. Uma delas é em relação a alta **dimensionalidade**, que é dada pelo tamanho do vocabulário. O problema disso é que, pela forma como o score é calculado, se uma palavra aparece n vezes em um documento, mas também está presente em todos os outros documentos, o peso TF-IDF dessa palavra será log(1) ou zero, Matematicamente, temos: [vide fórmula](https://github.com/wfaquieri/tf-idf_similarity_scores). Assim, os vetores da matriz TF-IDF são extremaentes **esparsos**, ou seja, possui a maioria das entradas igual a zero. Além disso, se temos duas frases (ou 2 documentos) que possuem exatamente o mesmo significado ou próximo, porém não utilizam as mesmas palavras, essa proximidade ou **relação semântica** entre os documentos não será capturada. Um exemplo, sejam 03 documentos: doc1 = "estou feliz", doc2 = "estou alegre" e doc3 = "estou triste". Esses 03 documentos receberiam por tf-idf o mesmo score, embora semanticamente há, claramente, mais similaridade entre o doc1 e o doc2. Em contraste, Doc2Vec é uma representação vetorial numérica de um objeto que captura algum tipo de semelhança semântica.
    
</font>
