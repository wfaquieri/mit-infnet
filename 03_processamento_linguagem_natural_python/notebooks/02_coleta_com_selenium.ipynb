```python
from selenium import webdriver 

chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument("--headless")
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")

wd = webdriver.Chrome('chromedriver', options=chrome_options)
wd.get("https://ge.globo.com/futebol/brasileirao-serie-a/")
```

```python
table_points = wd.find_element_by_class_name("tabela__pontos")
table_teams = wd.find_elements_by_class_name("classificacao__equipes--nome")
```

```python
import pandas as pd
from io import StringIO

content = StringIO(table_points.text)
table_points = pd.read_csv(content, sep=' ')
table_points.index = [t.text for t in table_teams]
del table_points['ÚLT.']
del table_points['JOGOS']
```
