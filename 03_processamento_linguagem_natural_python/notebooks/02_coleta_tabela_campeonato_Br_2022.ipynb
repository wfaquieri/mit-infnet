```python
import requests
from bs4 import BeautifulSoup
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}

url = "https://www.terra.com.br/esportes/futebol/brasileiro-serie-a/tabela/"
```

```python
page = requests.get(url, headers=headers)
```

```python
soup = BeautifulSoup(page.content, "html.parser")
```

```python
#print(soup.prettify())
```

```python
tables = soup.find_all("tbody")
```

```python
data = []
list_tr = tables[0].find_all("tr")
for tr in list_tr:
    team = {}
    list_td = tr.find_all("td")
    team['position'] = list_td[0].text
    team['name'] = list_td[2].a.attrs['title']
    team['points'] = list_td[4].text
    data.append(team)

```

```python
import pandas as pd
df  = pd.DataFrame(data)
df
```
